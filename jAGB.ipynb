```python
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u

# Project import
import artpop
```

```python
# artpop's matplotlib style
plt.style.use(artpop.mpl_style)
log_ages = artpop.MISTIsochrone._log_age_grid[40:-3]

# use this random state for reproducibility
rng = np.random.RandomState(112)
```

```python
ssp_old = artpop.MISTSSP(
    log_age = 10.1,          # log of age in years
    feh = -1,             # metallicity [Fe/H]
    phot_system = 'UBVRIplus', # photometric system(s)
    num_stars = 1e7,      # number of stars
    random_state = rng,   # random state for reproducibility
    ab_or_vega='vega'
)
RGB_o = ssp_old.select_phase('RGB')
```

```python
ssp_int = artpop.MISTSSP(
    log_age = 9.5,          # log of age in years
    feh = 0.5,             # metallicity [Fe/H]
    phot_system = 'UBVRIplus', # photometric system(s)
    num_stars = 1e7,      # number of stars
    random_state = rng,   # random state for reproducibility
    ab_or_vega='vega'
)
RGB_i = ssp_int.select_phase('RGB')
```

```python
ssp_young = artpop.MISTSSP(
    log_age = 8.5,          # log of age in years
    feh = 0,             # metallicity [Fe/H]
    phot_system = 'UBVRIplus', # photometric system(s)
    num_stars = 1e7,      # number of stars
    random_state = rng,   # random state for reproducibility
    ab_or_vega='vega'
)
RGB_y = ssp_young.select_phase('RGB')
```

```python
i_o = ssp_old.star_mags('Bessell_I')
g_o = ssp_old.star_mags('Bessell_V')
i_i = ssp_int.star_mags('Bessell_I')
g_i = ssp_int.star_mags('Bessell_V')
i_y = ssp_young.star_mags('Bessell_I')
g_y = ssp_young.star_mags('Bessell_V')
```

```python
plt.plot(g_i[RGB_i] - i_i[RGB_i], i_i[RGB_i], 'o',
c='tab:red', mec='k', label='intermediate')
plt.legend(loc='center right')
plt.minorticks_on()
plt.gca().invert_yaxis()
plt.xlabel(r'$g-i$')
plt.ylabel(r'M$_i$')
```

```python
AGB_o = ssp_old.select_phase('AGB')
AGB_i = ssp_int.select_phase('AGB')
AGB_y = ssp_young.select_phase('AGB')
```

```python
j_o = ssp_old.star_mags('2MASS_J')
k_o = ssp_old.star_mags('2MASS_Ks')
j_i = ssp_int.star_mags('2MASS_J')
k_i = ssp_int.star_mags('2MASS_Ks')
j_y = ssp_young.star_mags('2MASS_J')
k_y = ssp_young.star_mags('2MASS_Ks')
```

```python
plt.plot(j_o[RGB_o] - k_o[RGB_o], j_o[RGB_o], 'o',
         c='tab:red', mec='k', label='RGB old')
plt.plot(j_i[RGB_i] - k_i[RGB_i], j_i[RGB_i], 'o',
         c='tab:green', mec='k', label='RGB intermediate')
plt.plot(j_y[RGB_y] - k_y[RGB_y], j_y[RGB_y], 'o',
         c='tab:blue', mec='k', label='RGB young')
plt.plot(j_o[AGB_o] - k_o[AGB_o], j_o[AGB_o], 'o',
         c='tab:orange', mec='k', label='AGB old')
plt.plot(j_i[AGB_i] - k_i[AGB_i], j_i[AGB_i], 'o',
         c='tab:purple', mec='k', label='AGB intermediate')
plt.plot(j_y[AGB_y] - k_y[AGB_y], j_y[AGB_y], 'o',
         c='tab:pink', mec='k', label='AGB young')

# Adjust the legend location
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.minorticks_on()
plt.gca().invert_yaxis()
plt.xlabel(r'$j-k$')
plt.ylabel(r'M$_i$')
plt.axis([-1, 3, -2, -8])
```
